# About the data

Breast cancer is the most common cancer amongst women in the world. It accounts for 25% of all cancer cases and affected over 2.1 Million people in 2015 alone. It starts when cells in the breast begin to grow out of control. These cells usually form tumors that can be seen via X-ray or felt as lumps in the breast area.

## Exercise 1: Load the loan data (datasets/breast_cancer.csv) using pandas

```python
import pandas as pd
cancer = pd.read_csv("../../datasets/breast_cancer.csv", index_col=0)
cancer
```

```python
Target = cancer.diagnosis.values
cancer.drop("diagnosis", axis='columns', inplace=True)
cancer
```

## Exercise 2: Scale the data

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
cancer_scaler = scaler.fit_transform(cancer)
cancer_scaler
```

## Exercise 3: Perform KMeans clustering with 2 clusters

```python
from sklearn.cluster import KMeans

k=2
kmeans_cancer = KMeans(n_clusters= k, max_iter= 1000, n_init= 10)
kmeans_cancer.fit(cancer_scaler)
labels = kmeans_cancer.labels_
centroids = kmeans_cancer.cluster_centers_

crosstab = pd.crosstab(labels, cancer['diagnosis'], rownames=['Clusters'])
print(crosstab)
```

## Exercise 4: Use the elbow method to determine the number of clusters

```python
wcss = []
from matplotlib import pyplot as plt

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init = 'k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(cancer_scaler)
    wcss.append(kmeans.inertia_)

plt.figure(figsize = (15,5))
plt.plot(range(1,11), wcss, marker = 'o', linestyle = '--')
plt.title('Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Clusters')
plt.xticks(range(1,11))
plt.grid(True)

plt.show()
```

## Exercise 5: Plot the clusters and the centroids (color the data by cluster and then by the actual diagnosis)

```python
import numpy as np

for i in range(k):
    ds = cancer_scaler[np.where(labels==i)]
    plt.plot(ds[:,0],ds[:,1],'o')
    lines = plt.plot(centroids[i,0],centroids[i,1],'kx')
    plt.setp(lines,ms=10.0)
    plt.setp(lines,mew=2.0)
    
plt.show()
```
