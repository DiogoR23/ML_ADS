# Data Preprocessing Exercices

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
# load the wines dataset (datasets/wines.csv)
# add code here
wines = pd.read_csv("../../datasets/wines.csv", header = 0, sep = ",")
wines
```

```python
# check for missing values in the dataset
# add code here
wines.isna().sum()
```

```python
# if any, impute the missing values with the mean of the column
# add code here
wines2 = wines.copy()
wines2.fillna(wines2.median(), inplace=True)
wines2 = pd.DataFrame(wines2)
wines2.head(5)
```

```python
# scale the dataset using two different scaling methods
# add code here

# Standard Scalling
scaler = StandardScaler()
wines2_standard = scaler.fit_transform(wines2)
wines2_standard = pd.DataFrame(wines2_standard, columns = wines2.columns)
wines2_standard.head(5)
```

```python
# add code here

# Normalizer
normalizer = Normalizer()
wines2_normalizer = normalizer.fit_transform(wines2)
wines2_normalizer = pd.DataFrame(wines2_normalizer, columns = wines2.columns)
wines2_normalizer.head(5)
```

```python
# perform feature selection using two different methods
# add code here

# KBest
x = wines2_standard.drop('alcohol', axis = 1)
y = wines2_standard['alcohol']

kbest = SelectKBest(score_func = f_regression, k = 4)
data = kbest.fit_transform(x, y)

names = x.columns.values[kbest.get_support()]
data = pd.DataFrame(data, columns=names)
data
```

```python
# add code here

# Variance Threshold
print(wines2_normalizer.var())
x = wines2_normalizer.drop('alcohol', axis = 1)

selector = VarianceThreshold(threshold = 0.002)
data = selector.fit_transform(x)

names = x.columns.values[selector.get_support()]
data = pd.DataFrame(data, columns = names)
data
```
