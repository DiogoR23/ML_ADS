```python
# create a copy of the dataset and drop the rows with missing values
# add code here
df_wines_missing = df_wines.copy()
df_wines_missing = pd.DataFrame(df_wines_missing)
df_wines_missing.dropna(inplace=True)
```

```python
# create a copy of the dataset and fill the missing values with the median of the respective columns
# add code here
df_wines3 = df_wines.copy()
df_wines3.fillna(df_wines3.median(), inplace=True)
df_wines3.head(5)
```
